Finding High correlation Features

```python
def Colinearity(dataset,threshhold):
    col=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshhold:
                colname=corr_matrix.columns[i]
                col.add(colname)
    return col
```

```python
#Finding the columns which has High colinearity (0.85)

HighColinearityFeature=Colinearity(data[NumericFeatures],0.85)
HighColinearityFeature= list(HighColinearityFeature)
HighColinearityFeature
```

```python
data.drop(columns=HighColinearityFeature,inplace=True)
```

```python
# np.where(data['diagnosis']=='M',0,1)
data['diagnosis']=np.where(data['diagnosis']=='M',0,1)
data['diagnosis'].value_counts()
```

```python
x=data.drop(columns='diagnosis')
y=data.diagnosis
```

Spliting Test and Train data

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrainScaled=scaler.fit_transform(xtrain)
xtestScaled=scaler.transform(xtest)
```

Model Building

Best K value

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
score=[]
k_value=[i for i in range(1,31,2)]

for k in k_value:   
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrainScaled,ytrain)
    pred=knn.predict(xtestScaled)

    score.append(accuracy_score(ytest,pred)*100)

sns.lineplot(x = k_value, y = score, marker = 'o')
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

modelclasses=[
    ["K-NearestNeighbors",KNeighborsClassifier(n_neighbors=25)],
    ["DecisionTree",DecisionTreeClassifier()],
    ["RandomForest",RandomForestClassifier()],
    ["LogisticRegression",LogisticRegression()],
]
AccuracyScores=[]
# PrecisonScores=[]

for modelname,Model in modelclasses:        
        model=Model
        model.fit(xtrainScaled,ytrain)
        ypred=model.predict(xtestScaled)
        AccuracyScores.append((modelname,accuracy_score(ytest,ypred)*100))
        # PrecisonScores.append((modelname,precision_score(ytest,ypred)*100))

        print(f"--------------{modelname}----------------\n \n {classification_report(ytest,ypred)}")
```

```python
print("\n--------Accuracy Score-------------")
for mname,score in AccuracyScores:
        print("{} : {:.2f}".format(mname,score))
```

```python
plt.figure(figsize=(10,6))
AccuracyScores=pd.DataFrame(AccuracyScores,columns=['Model','Score'])
sns.countplot(data=AccuracyScores,x='Score',hue='Score',legend='brief',gap=.0)
```
